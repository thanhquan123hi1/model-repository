```python
from transformers import CLIPModel, CLIPProcessor
import os

# Create directory for CLIP model
clip_model_dir = "/content/clip-vit-large-patch14"
os.makedirs(clip_model_dir, exist_ok=True)

# Download and save CLIP model
print("Downloading CLIP ViT-L/14 model...")
clip_model = CLIPModel.from_pretrained("openai/clip-vit-large-patch14")
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-large-patch14")

# Save model
clip_model.save_pretrained(clip_model_dir)
clip_processor.save_pretrained(clip_model_dir)

print(f"✓ CLIP model saved to {clip_model_dir}")
```
