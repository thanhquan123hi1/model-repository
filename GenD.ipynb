```python
import os

source_root = "/content/data/Celeb-DF-v2"
target_root = "/content/deepfake-detection/datasets/CDFv2"

os.makedirs(target_root, exist_ok=True)

mapping = {
    "Celeb-real":       f"{source_root}/Celeb-real/frames",
    "Celeb-synthesis": f"{source_root}/Celeb-synthesis/frames",
    "YouTube-real":    f"{source_root}/YouTube-real/frames"
}

for name, src in mapping.items():
    dst = f"{target_root}/{name}"
    if os.path.exists(dst) or os.path.islink(dst):
        os.system(f"rm -rf {dst}")
    os.system(f"ln -s {src} {dst}")
    print(f"Linked {name}")
```
